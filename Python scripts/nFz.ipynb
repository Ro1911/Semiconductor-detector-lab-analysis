```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
from scipy.optimize import curve_fit
import statistics
from scipy import asarray as ar,exp
from scipy.signal import chirp, find_peaks, peak_widths
from sklearn.linear_model import LinearRegression
```

```python
E=[31,59.54,81,122]
Ch=[186.658,344.995,464.488,694.980]
Ch_err=[25.379,23.792,29.711, 19.431]
E_fine=np.linspace(30,125,100)
def fitfunc(x, a, c):
    return a+c*x
popt,pcov = curve_fit(fitfunc,E,Ch, sigma=Ch_err)
plt.plot(E_fine,fitfunc(E_fine,*popt),'r',label= r'Incline = {:1.4} $\pm$ {:1.2} Ch.no./E[keV]'.format(popt[1],np.sqrt(pcov[1][1])))
plt.errorbar(E,Ch, yerr=Ch_err, fmt='b.')
plt.title(r'Energy calibration for nFz diode')
plt.ylabel('Channel number')
plt.xlabel('Energy [keV]')
plt.legend(loc='best')
print('Incline = {:1.4} +/- {:1.2}'.format(popt[1],np.sqrt(pcov[1][1])))
#plt.savefig('C:/Users/RP/Documents/Uni_Helsinki/PAP339/Lab_project/Energy-calbration_nFz.pdf', format='pdf')
```
